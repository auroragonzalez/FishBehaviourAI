```python
from ultralytics import YOLO
```

## Video 1

```python
model = YOLO('yolov9c.pt')
results = model.train(data='/home/scaruso/DBS/prova01_train_videoUno-1/data.yaml',  # Path to data.yaml
                      epochs=50,                # Number of training epochs
                      imgsz=640,                # Image size for training
                      batch=16,                 # Batch size
                      name='yolo_trainingV2')     # Experiment name

metrics = model.val(data='/home/scaruso/DBS/prova01_train_videoUno-1/data.yaml',  # Path to data.yaml
                    imgsz=640,                # Image size for evaluation
                    batch=16,                  # Batch size
                    conf=0.5,
                    split="test")                 
print(metrics)
```

```python
model = YOLO('/home/scaruso/runs/final/Video1/TrainResults/yolo_training6/weights/best.pt')
metrics = model.val(data='/home/scaruso/DBS/prova01_train_videoUno-1/data.yaml',  # Path to data.yaml
                    imgsz=640,                # Image size for evaluation
                    batch=16,                  # Batch size
                    conf=0.5,
                    split="test")                 
print(metrics)
```
